## Prep data

To be able to create good prediction models the data has to be prepped in such a manner that the method recnognises it and can utilise to to gain insight into what is a proabable cause for the predicted class. The first task will be to predict the arrival time of the first unit sent and later discuss what are the biggest factors in that arrival time. Firstly the weather is added to enrich the dataset to see if thata can give beneficial parameters for the model.

```python
# Weither data and incidents
file = 'incidents_with_weather_fix.csv'
df = pd.read_csv(file,encoding='utf-8')
```

```python
# use data until 2017
df = df[df.DateOfCall <= "2016-12-31"]

# Lets see what columsn we have and choose the ones we want for our model
df.columns
```

```python
# THOSE are way to many and to many classes for some of them... look below

# Select columns to used for model
# select_columns = ['DateOfCall','HourOfCall','IncidentGroup','PropertyCategory',
#                   'PropertyType','IncGeo_BoroughCode','FirstPumpArriving_AttendanceTime',
#                  "Average Income","London zone",'Population','AttendanceTimeSeconds','wind_speed',
#                  'air_temperature','snow_depth',]

# df_selected = df[select_columns]
```

```python
# Select columns to be used for model
df_selected = df[['HourOfCall','PropertyCategory','IncGeo_BoroughCode','FirstPumpArriving_AttendanceTime',
                                'wind_speed','air_temperature','snow_depth']].copy()

# Change value to 0 if nan in snow_depth, and drop nan values in attendance
df_selected.dropna(subset = ['FirstPumpArriving_AttendanceTime'],inplace=True)
df_selected["snow_depth"] = df_selected["snow_depth"].fillna(0)
# df_selected["FirstPumpArriving_AttendanceTime"] = df_selected["FirstPumpArriving_AttendanceTime"].fillna(0)
```

```python
# Create dummy variables for Borough and PropertyCategory
df_selected = pd.get_dummies(df_selected, prefix="", prefix_sep='',columns=["IncGeo_BoroughCode"])
df_selected = pd.get_dummies(df_selected, prefix="", prefix_sep='',columns=["PropertyCategory"])
df_selected.head()
```

```python
# fix weather values to the mean of the column
df_selected['wind_speed'].fillna((df_selected['wind_speed'].mean()), inplace=True)
df_selected['air_temperature'].fillna((df_selected['air_temperature'].mean()), inplace=True)
df_selected.isnull().sum()
```

## Setup model

```python
# It takes about a 40 secs to train

from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation

#Split the dataset into features X and target variable Y.
X = df_selected.copy()
del X['FirstPumpArriving_AttendanceTime']
y = df_selected.FirstPumpArriving_AttendanceTime

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42) # 70% training and 30% test

regr = RandomForestRegressor(max_depth=4, random_state=0)

# Train the model on training data
regr.fit(X, y)
```

```python
# see how predicts on the test instance
y_pred = regr.predict(X_test)
```

```python
# Import functions to use for model performance
from sklearn.metrics import confusion_matrix, recall_score, accuracy_score, precision_score

# function to evaluate predictions
def evaluate(y_true, y_pred):
    # calculate and display confusion matrix
    labels = np.unique(y_true)
    cm = confusion_matrix(y_true, y_pred, labels=labels)
    print('Confusion matrix\n- x-axis is true labels \n- y-axis is predicted labels')
    print(cm)
    # calculate precision, recall, and F1 score
    accuracy = float(np.trace(cm)) / np.sum(cm)
    precision = precision_score(y_true, y_pred, average=None, labels=labels)[1]
    recall = recall_score(y_true, y_pred, average=None, labels=labels)[1]
    f1 = 2 * precision * recall / (precision + recall)
    print("accuracy:", accuracy)
    print("precision:", precision)
    print("recall:", recall)
    print("f1 score:", f1)
    
# function to evaluate predictions
def evaluate_continues(y_true, y_pred,regresson):
    # Manual calculation of R2, as (1 - u/v)
    u = ((y_true - y_pred) ** 2).sum() # u
    v = ((y_true - y_true.mean()) ** 2).sum() # 
    R2 = (1 - (u/v))
    gap = abs(np.asarray(y_pred) -  np.asarray(y_true)).sum() / np.asarray(y_test).sum()
    # with function
    print("The R2 score is: ",regresson.score(X_test,y_test))
    print("The R2 score manual is: ",R2)
    print("The % gap is: ",gap)
```

```python
evaluate_continues(y_test,y_pred,regr)
```

```python
# Check to see if we are overfitting
# Check how the model performs on the training set and compare to the results from the test data
y_train_pred = regr.predict(X_train)
evaluate_continues(y_train, y_train_pred,regr)
```
