To get a greater understanding of the dataset and its possibilities and limitation it was explored using standard methods. For categorical variables a bar chart was plotted of all the values, or the most common values where plotted if the number of unique values was very high. A histogram was plotted for the numerical variables to get a understanding of the distribution and explore if there were any outliers in the dataset.

All the three parts of the datasets where read into their own dataframe and then appended together to form a complete dataframe containing all the incident records.
Since the data is stored in an Excel file, the time it takes to read in the data is greater than if the data had been stored in a CSV file format. In order to do all the work with data in Python the fileformat was not modified in Excel beforehand.

```python
# Import necessary libraries and apply settings
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
from IPython.display import Image
%matplotlib inline
```

```python
# Read all the parts of the dataset into variables
df1 = pd.read_excel('IncidentData/LFB Incident data from January 2009 to December 2012.xlsx', index_col=None)
df2 = pd.read_excel('IncidentData/LFB Incident data from January 2013 to December 2016.xlsx', index_col=None)
df3 = pd.read_excel('IncidentData/LFB Incident data from January 2017.xlsx', index_col=None)
```

```python
# To demonstrate the differences in length the lengths of all the dataframes are printed
print("Number of columns in part 1: ", len(df1.columns))
print("Number of columns in part 2: ", len(df2.columns))
print("Number of columns in part 3: ", len(df3.columns))
```

```python
# Now the rows from part 1 and part 2 are appended to part 3
df = df3.append(df2, sort='False')
df = df.append(df1, sort = 'False')

# Reindex so that the columns are in teh same order as before
df = df.reindex(df3. columns, axis=1)

# Reset with our own index
df = df.reset_index(drop='True')

# Display a few rows at random
df.sample(n = 5, random_state = 42)
```

The next steps are to explore the distribution of each of the attribute and look for outliers or errors in the data. For convenience a function was created to make it easier to plot each attribute.
Since this dataset contains a lot of categorical variables with multiple values, no subplots were used since a greater size of the plots makes it easier to spot things that might be interesting to explore or potentially data errors.

```python
def simplePlottingFunction(dataframe_column, chart_type, top_values):
    plt.figure(figsize=(15,7))
    
    if (chart_type=='bar' and top_values == False):
        plt.bar(dataframe_column.value_counts().index, dataframe_column.value_counts().values)
    elif (chart_type=='scatter' and top_values == False):
        plt.scatter(dataframe_column.value_counts().index, dataframe_column.value_counts().values, s=1)
    elif (chart_type=='hist' and top_values == False):
        plt.hist(dataframe_column, 100)
    elif (chart_type=='bar' and top_values == True):
        plt.bar(dataframe_column.value_counts().index[0:20], dataframe_column.value_counts().values[0:20])
        plt.title(str(dataframe_column.name) + ' distribution - Top 20 values')
        plt.xlabel(str(dataframe_column.name))
        plt.ylabel('Incident Count')
        return
    plt.title(str(dataframe_column.name) + ' distribution')
    plt.xlabel(str(dataframe_column.name))
    plt.ylabel('Incident Count')
```
